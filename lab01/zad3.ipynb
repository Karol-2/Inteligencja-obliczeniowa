```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#dodanie bibliotek
```

```python
miasta = pd.read_csv("miasta.csv") #zaimportowanie pliku
print("zadanie 2 a")
print(miasta) # wyświetlenie zawartości pliku
print("wartości:")
print(miasta.values) # wyświetlenie wartości
```

```python
print("zadanie 2 b")
new_row = [2010, 460, 555, 405] # nowy rząd
miasta.loc[10] = new_row # dodanie go do miast

print(miasta) # wyświetlenie po zmianach
```

```python
print("zadanie 2 c")
plt.plot(miasta['Rok'], miasta['Gdansk'], 'ro-') #wykres dla Gdańska, czerwony z punktami

years = np.arange(miasta['Rok'].min(), miasta['Rok'].max() + 1, 10) #ładna podziałka na osi x
plt.xticks(years)

plt.title('Ludność Gdańska')
plt.xlabel('Rok') # tytuły i labele
plt.ylabel('Liczba ludności [w tys.]')
plt.grid()
plt.show() # wyświetlenie
```

```python
print("zadanie 2 d")
plt.plot(miasta['Rok'], miasta['Gdansk'], 'bo-', label='Gdańsk') #wykresy dla poszczególnych miast, z różnymi kolorami
plt.plot(miasta['Rok'], miasta['Poznan'], 'go-', label='Poznań')
plt.plot(miasta['Rok'], miasta['Szczecin'], 'ro-', label='Szczecin')

plt.xticks(years)
plt.title('Ludność w miastach Polski')
plt.xlabel('Rok')
plt.ylabel('Liczba ludności [w tys.]')
plt.grid()

plt.legend() # dodanie legendy

plt.show() # wyświetlenie
```

```python
print("zadanie 2 e")
kolumny = miasta[['Gdansk', 'Poznan', 'Szczecin']] # kolumny do zmian


def standaryzacja_kolumny(col):
    return (col - col.mean()) / col.std() # standaryzacja podanej kolumny według wzoru


miasta_standaryzowane = kolumny.apply(standaryzacja_kolumny, axis=0) # wykonanie standaryzacji dla kazdej kolumny

print(miasta_standaryzowane) # wyświetlenie

srednia = miasta_standaryzowane.mean() # średnia
odchylenie = miasta_standaryzowane.std() # odchylenie standardowe

print("ŚREDNIA:")
print(srednia)
print("ODCHYLENIE:")
print(odchylenie)
```

```python
print("zadanie 2 f")


def normalizacja_kolumny(col):
    return (col - col.min()) / (col.max() - col.min()) # normalizacja podanej kolumny według wzoru


miasta_normalizowane = kolumny.apply(normalizacja_kolumny, axis=0) # wykonanie normalizacji dla kazdej kolumny

print(miasta_normalizowane) # wyświetlenie

min = miasta_normalizowane.min() # minimalna wartość
max = miasta_normalizowane.max() # maksymalna wartość
print("MIN:")
print(min)
print("MAX:")
print(max)
```
